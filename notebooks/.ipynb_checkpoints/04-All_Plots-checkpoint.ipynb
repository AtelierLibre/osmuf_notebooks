import necessary libraries

```python
import osmuf as ouf
import osmnx as ox
```

select a place from a dictionary of test areas

```python
# dictionary of places of interest for convenience, 
places = {0:{'name':'buenos_aires', 'coordinates':(-34.5798,-58.4422),},
          1:{'name':'kaduna', 'coordinates':(10.53364, 7.37360),},
          2:{'name':'margery_street', 'coordinates':(51.5266,-0.1109),},
          3:{'name':'posadas', 'coordinates':(-27.37577,-55.90059),},
          4:{'name':'welwyn_garden_city', 'coordinates':(51.7938,-0.18645),},
          5:{'name':'hayes_lane', 'coordinates':(51.39425,-0.003),}
         }

# bounding box half-distance in meters
distance=500
```

```python
place=places[5]

place_name=place['name']
point=place['coordinates']
```

# 1. Collect the data

```python
study_area = ouf.study_area_from_point(point, distance)
net_city_blocks = ouf.places_from_point(point, distance)
buildings = ouf.buildings_from_gdf(net_city_blocks)
street_graph = ouf.street_graph_from_gdf(net_city_blocks)
```

### Pickle the data

```python
# insert pickle here
```

### Generate secondary data

```python
gross_city_blocks = ouf.gen_gross_city_blocks(street_graph, net_city_blocks)
gross_city_blocks.head(3)
```

### Project the data

```python
gross_city_blocks_proj = ox.project_gdf(gross_city_blocks)
net_city_blocks_proj = ox.project_gdf(net_city_blocks)
streets_proj = ouf.streets_proj_from_street_graph(street_graph)
buildings_proj = ox.project_gdf(buildings)
```

# 2. Process the projected data

```python
# generate the footprint and total GEA for each building
buildings_proj = ouf.measure_buildings(buildings_proj)
buildings_proj.head(3)
```

```python
# calculate the net_to_gross for the net_city_blocks
net_city_blocks_proj = ouf.gen_net_to_gross(net_city_blocks_proj, gross_city_blocks_proj)
```

```python
# calculate measures of the regularity of the net_city_blocks
net_city_blocks_form_factor = ouf.gen_regularity(net_city_blocks_proj)
net_city_blocks_form_factor.head(3)
```

### Link the buildings to the ids of the nearest net_city_block and street

```python
ax = net_city_blocks_proj.plot()
buildings_proj.plot(ax=ax, color='white')
streets_proj.plot(ax=ax, color='grey', linewidth=0.3)
```

```python
buildings_proj = ouf.join_buildings_place_id(buildings_proj, net_city_blocks_proj)
```

```python
buildings_proj = ouf.join_buildings_street_id(buildings_proj, streets_proj)
buildings_proj.head(3)
```

```python
net_city_blocks_proj = ouf.join_places_building_data(net_city_blocks_proj, buildings_proj)
net_city_blocks_proj.head()
```
